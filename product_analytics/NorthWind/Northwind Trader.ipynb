Goal 1: Rank Employee Sales Performance

```python
%%sql
WITH EmployeeSales AS(
        SELECT e.first_name || ' ' || e.last_name as employee_name,
               COUNT(*) AS num_sales, RANK() OVER win1
          FROM employees e
          JOIN orders o
            ON e.employee_id = o.employee_id
         GROUP BY e.employee_id
         WINDOW win1 AS(
            ORDER BY COUNT(*) DESC
         )
    )

SELECT *
  FROM EmployeeSales;
```

Goal 2: Identify Running Total of Monthly Sales

```python
%%sql
SELECT DATE_TRUNC('month', o.order_date)::DATE AS sale_month, 
      CAST(SUM(od.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10, 2)) AS monthly_sales,
      CAST(SUM(SUM(od.unit_price*od.quantity*(1-od.discount))) OVER(ORDER BY DATE_TRUNC('month', o.order_date)) AS NUMERIC(10, 2)) AS running_total
  FROM orders o 
  JOIN order_details od 
    ON o.order_id = od.order_id
 GROUP BY DATE_TRUNC('month', o.order_date)
```

Calculate Month-Over-Month Sales Growth

```python
%%sql
WITH MonthlySales AS(
    SELECT DATE_TRUNC('month', o.order_date)::DATE AS sale_month, 
      CAST(SUM(od.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10, 2)) AS monthly_sales
      FROM orders o 
      JOIN order_details od 
        ON o.order_id = od.order_id
     GROUP BY DATE_TRUNC('month', o.order_date)
),

PreMonthSales AS(
    SELECT EXTRACT('year' from sale_month) AS year, EXTRACT('month' from sale_month) AS month, monthly_sales, LAG(monthly_sales) OVER(ORDER BY sale_month) AS prev_month_sales
      FROM MonthlySales
)

SELECT *, CAST((monthly_sales - prev_month_sales)/prev_month_sales *100 AS NUMERIC(10,2)) AS month_over_month_growth
  FROM PreMonthSales
```

The data above indicates that sales for the food distribution company tend to increase during the winter months, particularly in December and January. This seasonal rise in sales could be attributed to holiday-related demand and the onset of colder weather, which often correlates with increased consumption of certain food products. On the other hand, the second quarter of the year, notably May and June, typically experiences a downturn in sales. This pattern could be related to the end of the school year and the beginning of warmer weather, which might influence changes in food purchasing patterns. Despite these observable trends, there is evidence of variability, with certain months showing unexpected dips in sales. Such fluctuations may be influenced by various factors, including but not limited to economic conditions, changing consumer preferences, or even unseasonal weather patterns, which can affect buying habits and consequently the company's monthly sales performance.

Goal 3: Identify High-Value Customers

```python
%%sql
WITH OrderValue AS(
    SELECT o.customer_id, o.order_id, CAST(SUM(od.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10,2)) AS total_order_values
      FROM orders o 
      JOIN order_details od
        ON o.order_id = od.order_id
     GROUP BY o.customer_id, o.order_id
),

ValueCategory AS(
  SELECT *, CASE WHEN total_order_values > AVG(total_order_values) OVER() THEN 'Above Average'
               ELSE 'Average/Below Average'
          END AS order_value
   FROM OrderValue
)

SELECT *
  FROM ValueCategory
LIMIT 10; 
```

```python
%%sql
-- counting how many orders are 'Above Average' for each customer.
WITH OrderValue AS(
    SELECT o.customer_id, o.order_id, CAST(SUM(od.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10,2)) AS total_order_values
      FROM orders o 
      JOIN order_details od
        ON o.order_id = od.order_id
     GROUP BY o.customer_id, o.order_id
),

ValueCategory AS(
  SELECT *, CASE WHEN total_order_values > AVG(total_order_values) OVER() THEN 'Above Average'
               ELSE 'Average/Below Average'
          END AS order_value
   FROM OrderValue
)

SELECT customer_id, COUNT(*) AS above_average_order_count
  FROM ValueCategory
 WHERE order_value = 'Above Average'
 GROUP BY customer_id
 LIMIT 10;
```

Goal 4: Examine Product and Categroy Performance

```python
%%sql
WITH ProductSales AS(
    SELECT c.category_id, c.category_name,
           CAST(SUM(p.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10,2)) AS category_sales
      FROM categories c
      JOIN products p 
        ON c.category_id = p.category_id
      JOIN order_details od
        ON p.product_id = od.product_id
      GROUP by c.category_id, c.category_name
)

SELECT *, CAST(category_sales / SUM(category_sales) OVER() * 100 AS NUMERIC(10,2)) AS percent_sales
  FROM ProductSales;
```

Beverages is the top category in terms of sales percentages, followed closely by Dairy Products. Produce and Grains/Cereals are the categories with the smallest sales percentage.

```python
%%sql
-- Find the top 3 products sold in each category
WITH ProductSales AS(
    SELECT p.category_id, p.product_id, p.product_name,
           CAST(SUM(p.unit_price*od.quantity*(1-od.discount)) AS NUMERIC(10,2)) AS product_sales
      FROM order_details od 
      JOIN products p
        ON od.product_id = p.product_id
     GROUP BY p.category_id, p.product_id, p.product_name
),

RankedProducts AS(
  SELECT *, ROW_NUMBER() OVER(PARTITION BY category_id ORDER BY product_sales DESC) AS row_number
    FROM ProductSales
)

SELECT *
  FROM RankedProducts
 WHERE row_number <= 3;

```
