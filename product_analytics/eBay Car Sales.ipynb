Find the invalid car registration years

```python
(~autos["registration_year"].between(1990,2016)).sum()/autos.shape[0]
```

Around 6.6% of our data has invalid car registration years and we may remove them.

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

It appears that most of the vehicles were first registered in the past 18 years since the date crawled.

Explore price by brand

```python
autos["brand"].value_counts(normalize = True)
```

As the top four brands constitute over 50% of the car brands in eBay, we choose them as common brands in the auction website.

```python
brand_counts = autos["brand"].value_counts(normalize = True)
common_brands = brand_counts[brand_counts > 0.09].index

autos[autos["brand"].isin(common_brands)]
```

```python
common_brands
```

```python
brand_price = {}
for brand in common_brands:
    avg_price = autos.loc[autos["brand"] == brand, "price"].mean()
    brand_price[brand] = avg_price
```

```python
brand_price
```

Among common brands, mercedes_benz and bmw have highest average price, around $8500, 1.5-3x the average prices of other common brands.

Store aggregate data in a dataframe

```python
bmp_series = pd.Series(brand_price)
```

```python
df = pd.DataFrame(bmp_series, columns = ['mean_price'])
```

```python
df
```

```python
brand_mileage = {}
for brand in common_brands: 
    avg_mileage = autos.loc[autos["brand"] == brand, "odometer_km"].mean()
    brand_mileage[brand] = avg_mileage
```

```python
df["brand_mileage"] = brand_mileage 
```

```python
df
```

There is a slight trend to the more expensive vehicles having higher mileage, with the less expensive vehicles having lower mileage. However, the average mileage among common brands did not vary much as the average price did, falling within 10% for top brands.
